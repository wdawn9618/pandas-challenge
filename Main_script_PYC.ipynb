```python
students_comp_data = pd.read_csv('students_complete.csv')
```

```python
schools_comp_data = pd.read_csv('schools_complete.csv')
```

```python
total_data = pd.merge(students_comp_data,schools_comp_data, how="left", on=["school_name", 'school_name'])
total_data.head()
```

```python
#total number of schools
Total_Schools = schools_comp_data['school_name'].unique()
All_Schools = len(Total_Schools)
print (All_Schools)
```

```python
#total number of students
All_Students = total_data["student_name"].count() 
print (All_Students)
```

```python
#total budget
All_Budget = schools_comp_data["budget"].sum()
print (All_Budget)
```

```python
#Average Math score
avg_math = total_data["math_score"].mean()
print (avg_math)
```

```python
#Average Reading score
avg_read = total_data["reading_score"].mean()
print (avg_read)
```

```python
#Percentage of students with math score - 70 or greater
#math_passing = total_data.loc[:,["math_score"]].head()
math_passing = (total_data[total_data['math_score'] >= 70].math_score.count())/All_Students
#students_passing_math = students_passing["Student ID"].count()
math_percent = math_passing * 100

print (math_percent)
```

```python
#Percentage of students with reading score - 70 or greater
read_passing = (total_data[total_data['reading_score'] >= 70].reading_score.count())/All_Students
read_percent = read_passing * 100
print (read_percent)
```

```python
#Percentage of students passing math and reading
total_stu_passing = total_data[(total_data['math_score'] >= 70) & (total_data['reading_score'] >= 70)]['Student ID'].count()/All_Students*100

print(total_stu_passing)
```

```python
#dataframe with all results
distr_summary = pd.DataFrame({"Total Schools": [All_Schools], "Total Students": [All_Students],
                              "Total Budget": [All_Budget], "Average Math Score": [avg_math],
                              "Average Reading Score": [avg_read], "% Passing Math": [math_percent],
                              "% Passing Reading": [read_percent], "% Overall Passing": [total_stu_passing]
                             })
print (distr_summary)                              
```

```python
sch_names_index = total_data.set_index('school_name').groupby(['school_name'])
sch_types =schools_comp_data.set_index('school_name')['type']
total_student = schools_comp_data.set_index('school_name')['size']
total_sch_budget = schools_comp_data.set_index('school_name')['budget']
per_stu_budget = total_sch_budget/total_student
average_math = sch_names_index['math_score'].mean()
average_reading = sch_names_index['reading_score'].mean()


 
# % who pass math at each school
School_math_pass = total_data[total_data['math_score'] >=70].groupby('school_name')['Student ID'].count()/total_student*100 

# % who pass reading at each school
School_read_pass = total_data[total_data['reading_score'] >=70].groupby('school_name')['Student ID'].count()/total_student*100 

# % who pass both at each school
All_passing = total_data[(total_data['reading_score'] >=70) & (total_data['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_student*100

#dataframe to store school data
Summary_school_table = pd.DataFrame({
    "School Type": sch_types,
    "Total Students": total_student,
    "Total School Budget": total_sch_budget,
    "Per Student Budget": per_stu_budget,
    "Average Math Score": average_math,
    "Average Reading Score": average_reading,
    "% Passing Math": School_math_pass ,
    "% Passing Reading": School_read_pass,
    "% Overall Passing": All_passing
})

Summary_school_table['Total School Budget'] = Summary_school_table['Total School Budget'].map('${:,.2f}'.format)
Summary_school_table['Per Student Budget'] = Summary_school_table['Per Student Budget'].map('${:,.2f}'.format)
Summary_school_table
```

```python
#Top performing schools
Pass_sort_top = Summary_school_table.sort_values(["% Overall Passing"], ascending=False)
Pass_sort_top.head()
```

```python
#Lowest performing schools 
Pass_sort_bottom = Summary_school_table.sort_values(["% Overall Passing"], ascending=True)
Pass_sort_bottom.head()
```

```python
#Math Scores by Grade
ninth_math = students_comp_data.loc[students_comp_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_math = students_comp_data.loc[students_comp_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_math = students_comp_data.loc[students_comp_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_math = students_comp_data.loc[students_comp_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

Math_by_grade = pd.DataFrame({
        "9th": ninth_math,
        "10th": tenth_math,
        "11th": eleventh_math,
        "12th": twelfth_math
})

Math_by_grade = Math_by_grade [['9th','10th','11th','12th']]
Math_by_grade
```

```python
#reading score by grade
ninth_reading = students_comp_data.loc[students_comp_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
tenth_reading = students_comp_data.loc[students_comp_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_reading = students_comp_data.loc[students_comp_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelfth_reading = students_comp_data.loc[students_comp_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

Reading_by_grade = pd.DataFrame({
        "9th": ninth_reading,
        "10th": tenth_reading,
        "11th": eleventh_reading,
        "12th": twelfth_reading
})

Reading_by_grade = Reading_by_grade [['9th','10th','11th','12th']]
Reading_by_grade
```

```python
#Scored by School Spending= per student
Sch_bins =[0, 585,629,644,675]
Bins_name = ["<$585", "$585-$629", "$630-$644", "$645-$675"]

##dataframe
Sum_schools= Summary_school_table.drop(columns=['Total Students','Per Student Budget','Total School Budget'])
Sum_schools["Spending Ranges (Per Student)"]=pd.cut(per_stu_budget,Sch_bins,labels=Bins_name,include_lowest=True)
Sum_schools
#groupby spending
Sum_sch_group = Sum_schools.groupby(["Spending Ranges (Per Student)"]).mean()
Sum_sch_group
```

```python
#Scores by School Size
Size_bins =[0, 1000, 2000, 5000 ]
Size_bins_name = ["Small(<1000)", "Medium(1000-2000)", "Large(2000-5000)"]
total_data['size'] = pd.cut(total_data['size'],Size_bins,labels=Size_bins_name)

size_sum= total_data.groupby('size')

average_math_score = size_sum['math_score'].mean()
average_reading_score =size_sum ['reading_score'].mean()

School_math_pass = total_data[total_data['math_score'] >=70].groupby('size')['Student ID'].count()/size_sum['Student ID'].count()*100 
School_read_pass = total_data[total_data['reading_score'] >=70].groupby('size')['Student ID'].count()/size_sum['Student ID'].count()*100 
All_passing = total_data[(total_data['reading_score'] >=70) & (total_data['math_score'] >= 70)].groupby('size')['Student ID'].count()/size_sum['Student ID'].count()*100 

scores_sch_size = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': math_percent,
    '% Passing Reading': read_percent,
    '% Overall Passing': All_passing
})
scores_sch_size.index.name = "Total Students"
scores_sch_size = scores_sch_size.reindex(Size_bins_name)
scores_sch_size
```
